# Data Retrieval

```python
{
 "fields": [{
     "numDimensions": 1024,
     "path": "embedding",
     "similarity": "cosine",
     "type": "vector"
   }]
}
```

```python
user_query = "What is the best horror movie to watch and why?"
```

```python
query_embedding = get_embedding(user_query)
```

```python
print(query_embedding)
```

```python
pipeline = [

    {
        "$vectorSearch": {
            "index": "vector_index",
            "queryVector": query_embedding,
            "path": "embedding",
            "numCandidates": 150,  # Number of candidate matches to consider
            "limit": 4,  # Return top 4 matches
        }
    },
    {
        "$project": {
            "fullplot": 1,  # Include the plot field
            "title": 1,  # Include the title field
            "genres": 1,  # Include the genres field
            "score": {"$meta": "vectorSearchScore"},  # Include the search score
        }
    }
]
```

```python
def vector_search(user_query,collection):

  query_embedding=get_embedding(user_query)
  print(query_embedding)

  if query_embedding is None:
    return "Invalid query or embeddig is failed"

  pipeline=[

            {
                "$vectorSearch":{

                "index": "vector_index",
                "queryVector": query_embedding,
                "path": "embedding",
                "numCandidates": 150,  # Number of candidate matches to consider
                "limit": 4,  # Return top 4 matches


                }

            },

              {
                 "$project":{

                "fullplot": 1,  # Include the plot field
                "title": 1,  # Include the title field
                "genres": 1,  # Include the genres field
                "score": {"$meta": "vectorSearchScore"},  # Include the search score
                 }

            }

           ]

  result=collection.aggregate(pipeline)
  return list(result)
```

```python
vector_search(user_query,collection)
```

```python
def get_search_result(query,collection):
  get_knowledge=vector_search(query,collection)
  search_result=""
  for result in get_knowledge:
    search_result += f"Title: {result.get('title','N/A')}, Plot: {result.get('fullplot','N/A')}"

  return search_result
```

```python
source_information = get_search_result(user_query,collection)
```

```python
combined_information = f"Query: {user_query}\nContinue to answer the query by using the Search Results:\n{source_information}."
print(combined_information)
```
